```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import urlparse
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from bokeh.io import output_notebook, show
output_notebook()
%matplotlib inline  
```

```python
def parsing(link, index, comment):
    r = urlopen(link).read()
    soup = BeautifulSoup(r, "lxml")
    content = soup.get_text()
    table = soup.find_all("table", class_="engineTable")
    arrange(table, index, comment)
```

```python
def arrange(table, index, comment):
    info = []
    
    # printing the top five runs scorers or wickets takers
    print(comment + ": ")
    for tr in table[0].find_all('tr')[1::2][0:5]:
        header = {}
        tds = tr.find_all('td')
        for i in range(len(tds)):
            header[i] = tds[i].text
            print(tds[i].text, end = ' ')
        info.append(header)
        print()
    plot(info, index, comment)
```

```python
def plot(info, index, comment):
    players = []
    score = [] #runs for batsmen while wicket for bowlers
    for i in range(5):
        players.append(info[i][0])
        score.append(int(info[i][index]))        
    y_pos = np.arange(0,2*len(players),2)

    dataframe(players, score, comment)
```

```python
def dataframe(players, scores, comment):
    run_wicket = comment.strip().split()[2] #runs or wickets as labels
    
    play = pd.DataFrame(players, columns=["Name"])
    run = pd.DataFrame(scores, columns=[run_wicket])

    df = pd.concat([play, run], axis=1)
    print(df)
    
    from bokeh.charts import Bar
    data = {
    'players': players,
    run_wicket: scores,
    }
    bar = Bar(data, label='players', values=run_wicket, agg='max', title=comment, tools='crosshair', legend=None)
    show(bar)
```

```python
batPath = 'http://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?id=11551;type=tournament'
bowlPath = 'http://stats.espncricinfo.com/ci/engine/records/bowling/most_wickets_career.html?id=11551;type=tournament'

parsing(bowlPath, 6, "Top 5 wickets takers")
parsing(batPath, 4, "Top 5 runs scorers")
```
